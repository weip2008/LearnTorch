To calculate and plot the Bollinger Bands for the `Close` column in a DataFrame, you can follow these steps:

1. **Calculate the moving average (MA)** of the `Close` column.
2. **Calculate the standard deviation (SD)** over a specified rolling window.
3. Use the formula for Bollinger Bands:
   - Upper Band = MA + (SD * 2)
   - Lower Band = MA - (SD * 2)

Hereâ€™s how to do it and plot the result using `matplotlib`:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace with your own data)
data = {
    'Close': [100, 102, 104, 105, 108, 107, 109, 170, 130,112, 15,128, 417, 116, 118, 120, 122, 118, 120, 122, 120, 122]
}
df = pd.DataFrame(data)

# Parameters for Bollinger Bands
window = 7  # Rolling window size
num_std = 1.1  # Number of standard deviations for the bands
alpha = 0.5 

# Calculate moving average (MA) and rolling standard deviation (SD)
df['MA'] = df['Close'].rolling(window=window).mean()
df['SD'] = df['Close'].rolling(window=window).std()
df['EWA'] = df['Close'].ewm(alpha=alpha, adjust=False).mean()

# Calculate upper and lower Bollinger Bands
df['Upper Band'] = df['EWA'] + (df['SD'] * num_std)
df['Lower Band'] = df['EWA'] - (df['SD'] * num_std)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Close'], label='Close Prices', color='blue')
plt.plot(df['MA'], label='Moving Average', color='green')
plt.plot(df['EWA'], label='Exponential Weighted Moving Average', color='black')
plt.plot(df['Upper Band'], label='Upper Bollinger Band', color='red')
plt.plot(df['Lower Band'], label='Lower Bollinger Band', color='red')

# Add labels and legend
plt.title('Bollinger Bands')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid()

# Show the plot
plt.show()
```


This code will calculate and plot the Bollinger Bands based on the `Close` prices in your DataFrame. You can adjust the `window` and `num_std` parameters to change the period and the width of the bands.
