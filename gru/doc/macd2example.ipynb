```python
# get data
import pandas as pd
from sqlalchemy import create_engine, inspect

# Create an SQLite engine
engine = create_engine('sqlite:///../../data/ESNQ_DB.db')

table_names = inspect(engine).get_table_names()
print(table_names)
ticker = 'ES5M'
df = pd.read_sql(ticker, engine)
```

```python
import pandas as pd
import pandas_ta as ta

# Create a new strategy with Stochastic RSI and MACD, with all time parameters multiplied by 5
strategy = ta.Strategy(
    name="ModifiedStrategy",
    ta=[
        {"kind": "stochrsi", "length": 70, "rsi_length": 70, "k": 35, "d": 35},  # Default 14 * 5
        {"kind": "macd", "fast": 12, "slow": 26, "signal": 9},  # Default (12, 26, 9)
        {"kind": "macd", "fast": 60, "slow": 130, "signal": 45}  # Default (12, 26, 9) * 5
    ]
)
```

```python
%timeit
# Apply the strategy to your dataframe
df.ta.strategy(strategy)
print(df.info())
```
