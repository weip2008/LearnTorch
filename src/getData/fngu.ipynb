The **FNGU** (MicroSectors FANG+ Index 3X Leveraged ETN) is an exchange-traded note (ETN) that provides leveraged exposure to the **FANG+ Index**. The FANG+ Index consists of ten high-growth tech-related companies, primarily from the technology and consumer discretionary sectors.

### Stocks included in the **FANG+ Index** (and thus indirectly in FNGU) are:

1. **Facebook (Meta Platforms)** - Ticker: **META**
2. **Apple** - Ticker: **AAPL**
3. **Amazon** - Ticker: **AMZN**
4. **Netflix** - Ticker: **NFLX**
5. **Google (Alphabet)** - Ticker: **GOOGL**
6. **Tesla** - Ticker: **TSLA**
7. **NVIDIA** - Ticker: **NVDA**
8. **Microsoft** - Ticker: **MSFT**
9. **Twitter (now X)** - Ticker: **X**
10. **Alibaba** - Ticker: **BABA**

These are the key stocks tracked by the FANG+ Index, and the FNGU ETN provides **3x leveraged exposure** to the daily performance of this index.

```python
import yfinance as yf
import pandas as pd
from sqlalchemy import create_engine

# Create an SQLite engine
engine = create_engine('sqlite:///data/fngu.db')

# List of stock symbols
fngus = ['META', 'AAPL', 'AMZN', 'NFLX', 'GOOGL', 'TSLA', 'NVDA', 'MSFT', 'X', 'BABA']
```

```python

# Loop through each stock symbol, download the data, and save it to the database
for stock in fngus:
    # Download the historical data for the stock
    df = yf.download(stock)

    # Save the data to the SQLite database
    df.to_sql(stock, engine, if_exists='replace')

    print(f"{stock} data saved to database.")
```
