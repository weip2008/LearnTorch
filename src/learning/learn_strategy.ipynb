The strategy
Market: S&P 500
Entry (buy on next days open):
1. S&P Price above 200-d Moving Average
2. 10-period RSI below 30
Exit (sell on next days open):
10-period RSI above 40 or after 10 trading days

```python
%timeit
import pandas as pd
import yfinance as yf
import ta
import numpy as np

df=yf.download( '^GSPC', start='1996-01-01')
df.describe
```

```python
df['sma200'] = ta.trend.sma_indicator(df.Close, window=200)

df['rsi'] = ta.momentum.rsi(df.Close,window=10)
```

```python
df['signal'] = np.where((df.Close > df.sma200) & (df.rsi < 30), True, False)
```

```python
df
```

```python
Buying_dates, Selling_dates = [],[]



for i in range(len(df)-1):
    if df.signal.iloc[i]:
        Buying_dates.append(df.iloc[i+1].name)
        for j in range(1,11):
            if df.rsi.iloc[i + j] > 40:
                Selling_dates.append(df.iloc[i + j + 1].name)
                break
            elif j == 10:
                Selling_dates.append(df.iloc[i + j + 1].name)
```

```python
Buying_dates = np.where(df.signal, df.index, np.nan)
```

```python
frame = pd.DataFrame({'Buying_dates':Buying_dates, 'Selling_dates':Selling_dates})
frame
```
